```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries.csv')

df
```

```python
sorted_by_names = df.sort_values(by='Country name')
sorted_by_names
```

```python
sorted_by_population = df.sort_values(by='Population', ascending=False)
sorted_by_population
```

```python
sorted_by_area = df.sort_values(by='Area (km^2)', ascending=False)
sorted_by_area
```

## Biggest countries by Population

```python
biggest_population_10 = sorted_by_population.head(10)
biggest_population_10
```

```python
biggest_population_10.plot(kind='bar', x='Country name', y='Population')
plt.xlabel('Country name')
plt.ylabel('Population')
plt.title('10 countries with the biggest population')
plt.show()
```

## Biggest countries by Area

```python
biggest_area_10 = sorted_by_area.head(10)
biggest_area_10
```

```python
biggest_area_10.plot(kind='bar', x='Country name', y='Area (km^2)')
plt.xlabel('Country name')
plt.ylabel('Area (km^2)')
plt.title('10 countries with the biggest area')
plt.show()
```

## Biggest by Population Density

```python
df['Population Density'] = df['Population'] / df['Area (km^2)']
df
```

```python
sorted_by_density = df.sort_values(by='Population Density', ascending=False)
sorted_by_density
```

```python
sorted_by_density.head(10).plot(kind='bar', x='Country name', y='Population Density')
plt.xlabel('Country name')
plt.ylabel('Population Density')
plt.title('10 countries with the biggest population density')
plt.show()
```

```python
df
```

```python
population_density = df['Population Density']
area = df['Area (km^2)']

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(area, population_density, color='orange', alpha=0.5)
plt.title('Scatter Plot of Population Density vs. Area')
plt.xlabel('Area (km^2)')
plt.ylabel('Population Density (people per km^2)')
plt.grid(True)
plt.show()
```

## Correlation Heatmap

```python
import seaborn as sns
# Assuming your DataFrame is named 'df'
# Select relevant columns for correlation calculation
correlation_data = df[['Population', 'Area (km^2)', 'Population Density']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
            fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap')
plt.show()
```

## Geospatial Visualization:

```python
import geopandas as gpd
# Load world shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge world dataframe with your dataframe 'df' based on country name
world = world.merge(df, how='left', left_on='name', right_on='Country name')

# Plot the world map with population density color-coded
fig, ax = plt.subplots(1, 1, figsize=(20, 15))
world.plot(column='Population Density', cmap='OrRd',
           linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('World Map with Population Density')
plt.show()
```
