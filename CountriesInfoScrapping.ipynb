```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = r'https://www.scrapethissite.com/pages/simple/'

request = requests.get(url)

soup = bs(request.text, 'html.parser')
```

```python
soup.prettify()
```

```python
countries = soup.find_all('div', class_='col-md-4')
print(countries)
```

```python
names = []
capitals = []
populations = []
areas = []
for country in countries:
    names.append(country.h3.text.strip())
    capitals.append(country.find_all('span')[0].text.strip())
    populations.append(country.find_all('span')[1].text.strip())
    areas.append(country.find_all('span')[2].text.strip())

print(names)
print(capitals)
print(populations)
print(areas)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Country name': names,
    'Capital': capitals,
    'Population': populations,
    'Area (km^2)': areas
})
```

```python
df
```

```python
df.to_csv('countries.csv', index=False)
```
